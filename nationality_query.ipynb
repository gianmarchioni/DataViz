```python
import pandas as pd
import plotly.express as px
import requests
from SPARQLWrapper import SPARQLWrapper, JSON
from tqdm.auto import tqdm  # Utilizzato per una barra di progressione
```

```python
# from google.colab import drive
# drive.mount('/content/drive')


# Function to  SPARQL
def get_nationality(artist_name):
    url = 'https://query.wikidata.org/sparql'
    query = """
    SELECT ?artist ?artistLabel ?countryLabel WHERE {
      ?artist rdfs:label "%s"@en.
      ?artist wdt:P27 ?country.
      SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
    } LIMIT 1
    """ % artist_name
    headers = {'User-Agent': 'Mozilla/5.0 (compatible; GoogleColab)', 'Accept': 'application/json'}
    response = requests.get(url, headers=headers, params={'query': query, 'format': 'json'})
    if response.status_code == 200:
        data = response.json()
        results = data.get('results', {}).get('bindings', [])
        if results:
            return results[0].get('countryLabel', {}).get('value')
    return None

# Percorso al file CSV nel tuo Google Drive
file_path = '/content/drive/My Drive/datavizproject/top_visited_2023.csv'
data = pd.read_csv(file_path)

# Applicare la funzione a ogni riga (potrebbe richiedere un po' di tempo a seconda del numero di artisti)
data['nationality'] = [get_nationality(name) for name in tqdm(data['name'], desc='Retrieving nationalities')]

# Percorso per il nuovo file CSV
output_file_path = 'artists_with_nationality.csv'
data.to_csv(output_file_path, index=False)

print("Nationalities added and file saved to your drive.")
```
