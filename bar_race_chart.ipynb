## Weekly data

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
weekly_views = df.groupby('article').resample('W').views.sum()
```

```python
weekly_views_df = weekly_views.reset_index()
```

```python
pivot_df = weekly_views_df.pivot(index='date', columns='article', values='views')
```

```python
pivot_df = pivot_df.fillna(0).cumsum()
```

```python
import pandas as pd
import bar_chart_race as bcr

bcr.bar_chart_race(
    df=pivot_df,
    filename='viewsdata.mp4',
    orientation='h',
    sort='desc',
    n_bars=10,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                  's': f'Total Views: {v.nlargest(6).sum():,.0f}',
                                  'ha': 'right', 'size': 8, 'family': 'DejaVu Sans'},
    period_length=700,
    figsize=(12, 6),
    dpi=300,
    cmap='dark12',
    title='Views Over Time per Artist',
    bar_label_size=7,
    tick_label_size=7,
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)

# The chart will be saved as 'viewsdata.mp4' in the specified path
from google.colab import files

files.download('viewsdata.mp4')
```
